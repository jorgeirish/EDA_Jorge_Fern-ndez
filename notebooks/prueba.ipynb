```python
#Para la manipulación de datos
import pandas as pd

#Servicio y driver de Chrome de Selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

#Las opciones que vamos a tener para buscar elementos
from selenium.webdriver.common.by import By

#Para cuando queramos mandar pulsaciones de teclado
from selenium.webdriver.common.keys import Keys

#Hacemos que espere
import time
```

```python
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
```

```python
# Búsqueda de los datos a través de selenium

url = "https://www.juntadeandalucia.es/medioambiente/atmosfera/informes_siva/cuantitativo/2022/"
driver.get(url)
```

```python
search = driver.find_elements(By.TAG_NAME, "a")
search

# Descarcar cada uno de los csv de la página

for i in search[1:]:
    i.click()
```
