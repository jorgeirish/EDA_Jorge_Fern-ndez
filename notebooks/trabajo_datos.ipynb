```python
temp_and = pd.read_csv("AEMET/aemet_and.csv")
temp_and.head()
```

```python
out_nan2 = temp_and.dropna(how="any")
out_nan2.head()
```

```python
out_nan2.info()
```

```python
out_nan2.drop("Unnamed: 0", axis = 1, inplace=True)
out_nan2.info()
temp = out_nan2.reindex(columns=["fecha", "nombre", "provincia", "altitud", "tmed", "tmin", "tmax", "prec", "velmedia", "racha"]).sort_values(["nombre", "fecha"])
```

```python
temp.head()
temp.isnull().any()
```

```python
temp["nombre"].unique()
```

```python
#Sacar las capitales de provincia
temp_cap = temp[(temp["nombre"]=="ALMERÍA AEROPUERTO")|
                (temp["nombre"]=="CÁDIZ")|
                (temp["nombre"]=="CÓRDOBA AEROPUERTO")|
                (temp["nombre"]=="GRANADA AEROPUERTO")|
                (temp["nombre"]=="HUELVA")|
                (temp["nombre"]=="JAÉN")|
                (temp["nombre"]=="MÁLAGA")|
                (temp["nombre"]=="SEVILLA AEROPUERTO")]
temp_cap.head()
#temp_cap.isnull().any()
#temp_cap.info()
```

```python
temp_cap["fecha"] = pd.to_datetime(temp_cap["fecha"])
temp_cap.info()
```

```python
nocomas = temp_cap.replace(",",".", regex=True)
noip = nocomas.replace("Ip", None, regex=True)
temp_f= noip.dropna(how="any")
temp_f
```

```python
temp_f["tmed"]=temp_f["tmed"].astype(float)
temp_f["tmax"]=temp_f["tmax"].astype(float)
temp_f["tmin"]=temp_f["tmin"].astype(float)
temp_f["prec"]=temp_f["prec"].astype(float)
temp_f["velmedia"]=temp_f["velmedia"].astype(float)
temp_f["racha"]=temp_f["racha"].astype(float)
temp_f.info()
```

```python
temp_f["mes"] = temp_f["fecha"].dt.month
temp_f = temp_f.reindex(columns=["fecha", "mes", "nombre", "provincia", "altitud", "tmed", "tmin", "tmax", "prec", "velmedia", "racha"]).sort_values(["nombre", "fecha"])
temp_f.info()
```
